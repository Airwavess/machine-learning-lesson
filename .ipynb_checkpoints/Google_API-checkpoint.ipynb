```python
ids = []
for city in cities:
    geocode_result = gmaps.geocode(city)
    location = geocode_result[0]['geometry']['location']
    store = gmaps.places_radar(keyword='寵物', location=location, radius=2500)['results']
    print('以{}為中心半徑2500公尺的寵物店家數量: {}'.format(city, len(store)))
    
    for place in store:
        ids.append(place['place_id'])
```

```python
gmaps.place(place_id='ChIJgQMaaTipQjQR6XYl3SPMFxA', language='zh-TW')
```

```python
stores_info = []
id = list(set(ids))
for i, id in enumerate(ids):
    if i%100 == 0:
        print('progress: {} of {}'.format(i, len(ids)))
    stores_info.append(gmaps.place(place_id=id, language='zh-TW')['result'])
```
